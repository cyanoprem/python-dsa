```python
class ListNode:
  def __init__(self, value=0, next=None):
      self.value = value
      self.next = next
```

```python
# Helper function to print list nodes
def print_list(node):
    while node:
        print(node.value, end=" -> ")
        node = node.next
    print("None")
```

```python
# Create a list: 1 -> 2 -> 3 -> 4 -> 5 -> None
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
```

```python
print_list(head)  # Expected output: 1 -> 2 -> 3 -> 4 -> 5 -> None
```

```python
def reverse_list(head):
    prev = None
    current = head
    while current:
        next_node = current.next  # Store next node
        current.next = prev       # Reverse current node's pointer
        prev = current            # Move pointers one position ahead
        current = next_node
    head = prev
    return head
```

```python
# Reverse the list
head = reverse_list(head)
```

```python
print_list(head)  # Expected output: 5 -> 4 -> 3 -> 2 -> 1 -> None
```

```python
# Reverse the list back
head = reverse_list(head)
```

```python
print_list(head)  # Expected output: 1 -> 2 -> 3 -> 4 -> 5 -> None
```

```python
# Helper function to create a loop
def create_loop(head, loop_start_pos):
    if loop_start_pos is None:
        return
    loop_start_node = last_node = head
    for _ in range(loop_start_pos):
        loop_start_node = loop_start_node.next
    while last_node.next:
        last_node = last_node.next
    last_node.next = loop_start_node
```

```python
# Introduce a loop: 5 -> 3 (back to the third node)
create_loop(head, 2)
```

```python
print_list(head)  # Expected output: 1 -> 2 -> 3 -> 4 -> 5 -> None
```

```python
def detect_and_remove_loop(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            remove_loop(head, slow)
            return True
    return False

def remove_loop(head, loop_node):
    ptr1 = head
    while True:
        ptr2 = loop_node
        while ptr2.next != loop_node and ptr2.next != ptr1:
            ptr2 = ptr2.next
        if ptr2.next == ptr1:
            break
        ptr1 = ptr1.next
    ptr2.next = None
```

```python
# Detect and remove the loop
detect_and_remove_loop(head)
print("Loop removed.")
```

```python
print_list(head)  # Expected output: 5 -> 4 -> 3 -> 2 -> 1 -> None
```

```python
def add_at_start(head, value):
    new_node = ListNode(value)
    new_node.next = head
    head = new_node
    return head
```

```python
def remove_from_start(head):
    if head is not None:
        head = head.next
    return head
```

```python
# Add an element at the start:
head = add_at_start(head, 0)
print("After adding at start:")
print_list(head)  # Expected output: 0 -> 1 -> 2 -> 3 -> 4 -> 5 -> None
```

```python
# Remove the first element:
head = remove_from_start(head)
print("After removing from start:")
print_list(head)  # Expected output: 1 -> 2 -> 3 -> 4 -> 5 -> None
```
